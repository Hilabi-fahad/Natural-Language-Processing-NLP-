<h4>Tokenization</h4>
<p>Breaks a complex sentence into words</p>

```python
from nltk.tokenize import word_tokenize
```

```python
AI_tokens= word_tokenize(AI)
AI_tokens
```

```python
len(AI_tokens)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

<h6>Word count in the paragraph</h6>

```python
for word in AI_tokens:
    fdist[word.lower()]+=1
fdist
```

```python
top10 = fdist.most_common(10)
top10
```

<h4>Key terms of Tokenization</h4>
<ul>
<li>bigrams : Tokens of 2 number of consecutive written words</li>
<li>trigrams: Tokens of 3 number of consecutive written words</li>
<li>ngrams: Tokens of any number of consecutive written words</li>
</ul>

```python
from nltk.util import bigrams , trigrams ,ngrams
```

```python
string ="the best and most beautiful things in the world cannot been seen or even touched , they can felt with the heart "
quotes = nltk.word_tokenize(string)
quotes
```

```python
quotes_bigrams =list(nltk.bigrams(quotes))
quotes_bigrams
```

```python
quotes_bigrams =list(nltk.trigrams(quotes))
quotes_bigrams
```

```python
quotes_ngrams =list(nltk.ngrams(quotes,6))
quotes_ngrams
```
